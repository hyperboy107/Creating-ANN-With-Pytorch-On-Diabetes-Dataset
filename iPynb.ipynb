```python
X=df.drop('Outcome',axis=1).values### independent features #.values mean converting to numpy value
y=df['Outcome'].values###dependent features
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
y_train
```

```python
pip install torch
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
```

```python
# creating tensors
X_train=torch.FloatTensor(X_train)
X_test=torch.FloatTensor(X_test)
y_train=torch.LongTensor(y_train)
y_test=torch.LongTensor(y_test)
```

```python
df.shape
```

```python
X_train.device
```

```python
#creating model
# Forwording input to hideen1 & 2 to the output
class ANN_Model(nn.Module):
    def __init__(self,input_features=8,hidden1=20,hidden2=20,out_features=2):
        super().__init__()
        self.f_connected1=nn.Linear(input_features,hidden1)
        self.f_connected2=nn.Linear(hidden1,hidden2)
        self.out=nn.Linear(hidden2,out_features)
    def forward(self,x):
        x=F.relu(self.f_connected1(x))
        x=F.relu(self.f_connected2(x))
        x=self.out(x)
        return x
```

```python
torch.manual_seed(20)
model=ANN_Model()
```

```python
model.parameters
```

```python
model.parameters()
```

```python
for i in model.parameters():
    print(i.is_cpu)
```

```python
model=model.cpu()
```

```python
# bckword loss_finction
loss_function=nn.CrossEntropyLoss()
optimizer=torch.optim.Adam(model.parameters(),lr=0.01)
```

```python
import time
start_time=time.time()
epochs=500
final_losses=[]
for i in range(epochs):
    i=i+1
    y_pred=model.forward(X_train)
    loss=loss_function(y_pred,y_train)
    final_losses.append(loss)
    if i%10==1:
        print("Epoch number: {} and the loss : {}".format(i,loss.item()))
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
print(time.time()-start_time)
```

```python
pip install matplotlib
```

```python
### plot the loss function
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#### Prediction In X_test data
predictions=[]
with torch.no_grad(): #For removing gradients
    for i,data in enumerate(X_test):
        y_pred=model(data)
        predictions.append(y_pred.argmax().item())
        print(y_pred.argmax().item())
```

```python
predictions
```

```python
#Comparing predictions with x_test data we use confusion matrix
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,predictions)
cm
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(cm,annot=True)
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
```

```python
from sklearn.metrics import accuracy_score
score=accuracy_score(y_test,predictions)
score
```
